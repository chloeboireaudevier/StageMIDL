````python
import pandas as pd
import numpy as np
import json
````

````python
train_csv = pd.read_csv("SousCorpusMessagesRecap.csv",encoding='latin-1')
train_csv.rename(columns = {train_csv.columns[0] : "messageID"},inplace=True)
train_csv.head()
````

````python
# Contenu mask
mask_contenu = (
    (train_csv["Contenu Jose"] == train_csv["Contenu Chloe"]) &
    (train_csv["Contenu Jose"] == train_csv["Contenu Sebastien"]) &
    (train_csv["Contenu Jose"] == train_csv["Contenu Julien"])
)

# Forme mask
mask_forme = (
    (train_csv["Forme Jose"] == train_csv["Forme Chloe"]) &
    (train_csv["Forme Jose"] == train_csv["Forme Sebastien"]) &
    (train_csv["Forme Jose"] == train_csv["Forme Julien"])
)

# Nature mask
mask_nature = (
    (train_csv["Nature Jose"] == train_csv["Nature Chloe"]) &
    (train_csv["Nature Jose"] == train_csv["Nature Sebastien"]) &
    (train_csv["Nature Jose"] == train_csv["Nature Julien"])
)

mask = mask_contenu & mask_forme & mask_nature

filtered = train_csv[mask]
filtered
````

````python
def collapse_columns(df, cols, new_name):
    """
    Collapse a group of identical columns into a single column.

    Parameters
    ----------
    df : pd.DataFrame
        The dataframe to operate on.
    cols : list of str
        The list of column names that should be identical.
    new_name : str
        The name of the resulting collapsed column.

    Returns
    -------
    pd.DataFrame
        A new dataframe with the collapsed column.
    """
    df = df.copy()

    # Create the new column using the first column (they are identical)
    df[new_name] = df[cols[0]]

    # Drop the original columns
    df = df.drop(columns=cols)

    return df
````

````python
filtered_contenu = train_csv[mask_contenu]
cols_contenu = ["Contenu Jose", "Contenu Chloe", "Contenu Sebastien", "Contenu Julien"]

filtered_contenu = train_csv[mask_contenu]
filtered_contenu = collapse_columns(filtered_contenu, cols_contenu, "contenu")

filtered_contenu
````

````python
filtered_nature = train_csv[mask_nature]
cols_nature = ["Nature Jose", "Nature Chloe", "Nature Sebastien", "Nature Julien"]

filtered_nature = train_csv[mask_nature]
filtered_nature = collapse_columns(filtered_nature, cols_nature, "nature")

filtered_nature
````

````python
filtered_forme = train_csv[mask_forme]
cols_forme = ["Forme Jose", "Forme Chloe", "Forme Sebastien", "Forme Julien"]

filtered_forme = train_csv[mask_forme]
filtered_forme = collapse_columns(filtered_forme, cols_forme, "forme")

filtered_forme
````

````python
def get_prompts(df,name):
    # Prompt_agent
    # Group by message_num and build the desired structure
    result = {
        "annotations": []
    }

    for message_num, group in df.groupby("messageID"):
        utterances = group.apply(lambda row: {
            name: row[name]
        }, axis=1).tolist()

        result["annotations"].append({
            "message_num": int(message_num),
            "utterances": utterances
        })

    # Export to .txt file with ```json annotation
    with open("prompts_by_category/prompt_agent_from_corpus_"+name+".txt", "w", encoding="utf-8") as f:
        f.write("```json\n")
        json.dump(result, f, indent=2, ensure_ascii=False)
        f.write("\n```")

    # Prompt_user
    # Convert to desired JSON structure
    conversation_data = {
        "seance_ID": 1,
        "groupe_ID": 4,
        "conversation": []
    }

    # Sort by message_num for logical order (optional)
    df_sorted = df.sort_values("messageID")

    # Build conversation entries
    for _, row in df_sorted.iterrows():
        conversation_data["conversation"].append({
            "messageID": int(row["messageID"]),
            "role": row["role"],
            "message": row["message"]
        })

    # Write to .txt file with ```json annotation
    with open("prompts_by_category/prompt_user_from_corpus_"+name+".txt", "w", encoding="utf-8") as f:
        #f.write("```json\n")
        json.dump(conversation_data, f, indent=2, ensure_ascii=False)
        #f.write("\n```")

````

````python
get_prompts(filtered_forme,"forme")
get_prompts(filtered_nature,"nature")
get_prompts(filtered_contenu,"contenu")
````

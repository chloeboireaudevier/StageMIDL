````python
import pandas as pd
import numpy as np
import json
````

````python
train_csv = pd.read_csv("SousCorpusMessagesTrainPrompt.csv",encoding='latin-1')
train_csv.head()
````

````python
# Prompt_agent
# Group by message_num and build the desired structure
result = {
    "annotations": []
}

for message_num, group in train_csv.groupby("messageID"):
    utterances = group.apply(lambda row: {
        "text": row["message"],
        "form": row["form"],
        "tone": row["tone"],
        "content": row["content"],
        "nature": row["nature"]
    }, axis=1).tolist()

    result["annotations"].append({
        "message_num": int(message_num),
        "utterances": utterances
    })

# Export to .txt file with ```json annotation
with open("prompt_agent_from_corpus.txt", "w", encoding="utf-8") as f:
    f.write("```json\n")
    json.dump(result, f, indent=2, ensure_ascii=False)
````

````python
# Prompt_user
# Convert to desired JSON structure
conversation_data = {
    "seance_ID": 1,
    "groupe_ID": 4,
    "conversation": []
}

# Sort by message_num for logical order (optional)
df_sorted = train_csv.sort_values("messageID")

# Build conversation entries
for _, row in df_sorted.iterrows():
    conversation_data["conversation"].append({
        "messageID": int(row["messageID"]),
        "role": row["role"],
        "message": row["message"]
    })

# Write to .txt file with ```json annotation
with open("prompt_user_from_corpus.txt", "w", encoding="utf-8") as f:
    f.write("```json\n")
    json.dump(conversation_data, f, indent=2, ensure_ascii=False)
    f.write("\n```")
````
